# program for increment and decrement

```
#include<stdio.h>
int main()
{
	int a,b;
	a=10;
	b=0;
	printf("\n a=%d b=%d",a,b);
	b=a;
	printf("\n a=%d b=%d",a,b);
    b=++a;
    printf("\n a=%d b=%d",a,b);
    b=--a;
    printf("\n a=%d b=%d",a,b);
    b=a++;
    printf("\n a=%d b=%d",a,b);
    b=a--;
    printf("\n a=%d b=%d",a,b);
    return 0;
}
```

# asciii values

```
#include<stdio.h>
int main()
{
	char x;
	scanf("%c",&x);
	if(x>65 && x<=90)
	{
		printf("upper case");
	
	}
	else if(x>=97&&x<=122)
	{
		printf("lower case");
		
	}
	else if(x>=44&&x<=57)
	{
		printf("digit");
	}
	return 0;
}
```

# vowels and consonents

```
#include<stdio.h>
int main()
{
	char x;
	scanf("%c",&x);
	if(x=='a'|| x=='e'||x=='i'||x=='o'||x=='u')
	{
		printf("vowle");
	
	}
	else 
	{
		printf("consonents");
	}
	return 0;
}
```

# switch case

```
# include<stdio.h>
int main()
{
	int a;
	scanf("%d",&a);
	switch(a)
	{
	
		case 1:
			
			printf("sunday");
		
		    break;
		case 2:
			
			printf("monday");
		
		break;
		case 3:
		
			printf("tuseday");
		
		break;
		case 4:
			
			printf("wednesday");
		
		break;
		case 5:
			
			printf("thursday");
		
		break;
		case 6:
			
			printf("friday");
		
		break;
		case 7:
			
				printf("saturday");
			
			break;
      }
		return 0;
		
			
}
```

# swhitch case

```
#include<stdio.h>
int main()
{
	char a;
	scanf("%c",&a);
	switch(a)
	{
		case 'a':
			printf("vowle");
		break;
		case 'e':
			printf("vowle");
		break;
		case 'i':
			printf("vowle");
		break;
		case 'o':
			printf("vowle");
		break;
		case 'u':
			printf("vowle");
		break;
		default : printf("consonent");
	}
	return 0;
}
```
