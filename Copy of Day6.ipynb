# for loop syntax
for x in range(parameters):
statements

```
def printseries(lb,ub):
    for x in range(lb,ub+1):
        print(x,end=" ")
    return
printseries(1,10)
```

# Alternative numbers 

```
def printseries(lb,ub):
    for x in range(lb,ub+1):
        if (x%2==0):
            print(x,end=" ")
    return
printseries(500,520)
```

# jump statement

```
for x in range(500,520+1,2):
    print(x,end=" ")
```

```
for x in range(500,520+1):
     if (x%2==0):
         print(x,end=" ")
```
